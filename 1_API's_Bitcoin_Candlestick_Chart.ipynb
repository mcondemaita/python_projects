```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import plotly.graph_objects as go

!pip install plotly

from datetime import datetime, timezone
```

```python
!pip install pycoingecko
from pycoingecko import CoinGeckoAPI
cg= CoinGeckoAPI()
bitcoin_data = cg.get_coin_market_chart_by_id(id="bitcoin", vs_currency='usd', days=30) #funtion to request data 
```

```python
bitcoin_price_data = bitcoin_data['prices'] #only select what we need
```

```python
data = pd.DataFrame(bitcoin_price_data, columns=['TimeStamp', 'Price'])
data
```

```python
#fix TimeStamp 
data['Date'] = pd.to_datetime(data['TimeStamp'], unit = 'ms') #output to miloseconds 
data['Date']
```

```python
data #we append the date to our data 
```

```python
#represent on a candlestick data 
candlestick_data = data.groupby(data.Date.dt.date).agg({'Price': ['min', 'max', 'first', 'last']}) #into a dictionary 
candlestick_data.head()
```

```python
#plug them in a figure 
fig = go.Figure(data =[go.Candlestick(x= candlestick_data.index, 
                open= candlestick_data['Price']['first'],
                high = candlestick_data['Price']['max'],
                low = candlestick_data['Price']['min'],
                close = candlestick_data['Price']['last'])
                ])
fig.update_layout(xaxis_rangeslider_visible=False, xaxis_title ='Date', 
                  yaxis_title='Price USD $', title = 'Bitcoin Cadlestick chart Over the 30 Days')

#plot(fig, filename= 'bitcoin_candlestick_graph.html')
```
