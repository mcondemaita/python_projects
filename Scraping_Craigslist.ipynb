```python
#Import modules and funtions
import requests
from bs4 import BeautifulSoup # This module helps in web Scrapping
import requests #this module helps us download a web page 
```

```python
r = requests.get("https://newyork.craigslist.org/search/lbs?query=contractors&bundleDuplicates=1&searchNearby=2")
```

```python
r.status_code #request is saccesfull 
```

```python
r.headers["Content-Type"]
```

```python
html=r.text
html
```

```python
soup = BeautifulSoup(html, "html5lib")
soup
```

```python
#tag the HTML 
tag_object = soup.title
tag_object
```

```python
tag_object = soup.h3 #only the 1st one is selected
tag_object
```

```python
tag_child = tag_object.a #to navegate the header to the child
tag_child
```

```python
tag_child.attrs #tag attributes will provide the information 
```

```python
tag_child.string #get the string 
```

```python
parent_tag = tag_child.parent #to navegate back to the parent 
parent_tag
```
